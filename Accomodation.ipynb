```python
calendar['date'] = pd.to_datetime(calendar['date'], format = "%Y-%m-%d")
calendar['day'] = calendar['date'].dt.day
calendar['month'] = calendar['date'].dt.month
calendar['Year'] = calendar['date'].dt.year
calendar['day_name'] = calendar['date'].dt.strftime('%A')
calendar['month_name'] = calendar['date'].dt.strftime('%B')
```

```python
calendar['available'].replace('t','1', inplace = True)
calendar['available'].replace('f','0', inplace =True)
```

```python
calendar.head()
```

```python
# count of Listing based on each month.

a_listing = calendar.groupby(['month_name'])['listing_id'].count().sort_values(ascending=False).to_frame().reset_index()
a_listing = a_listing.rename(columns = {'month_name' : 'Months','listing_id' : 'Number of Listings'})
a_listing
```

```python
# we will use autopct to indicate percent lables on pie chart

from matplotlib import pyplot as plt
plt.pie(a_listing['Number of Listings'], labels = a_listing['Months'], autopct = '%.2f%%', pctdistance = 0.6,radius=3)
plt.rcParams['font.size'] = 15
plt.show()
```

```python
# Number of listings were published in every Year.

b_listing = calendar.groupby(['Year'])['listing_id'].count().sort_values(ascending=False).to_frame().reset_index()
b_listing = b_listing.rename(columns = {'listing_id' : 'Number of Listings'})
b_listing
```

```python
from matplotlib import pyplot as plt
plt.pie(b_listing['Number of Listings'], labels = b_listing['Year'], autopct = '%.2f%%', pctdistance = 0.6,radius=3)
plt.rcParams['font.size'] = 15
plt.show()
```

```python
# Number of listings were published in every Year and in each month.

c_listing = calendar.groupby(['Year','month_name'])['listing_id'].count().to_frame().reset_index()
c_listing = c_listing.rename(columns = {'month_name' : 'Month','listing_id' : 'Number of Listings'})
c_listing
```

```python
import seaborn as sb
plt.figure(figsize = (12,8))
sb.countplot(y = "month_name", hue = "Year", data = calendar)
```

```python
listings_merge = pd.merge(listings,listings_brief,left_on = "id",right_on = "id")
listings_merge.head()
```

```python
# Top 5 hosts where the maximum number of listings are published

d_listing = listings_merge.groupby(['host_name_x'])['id'].count().sort_values(ascending=False).to_frame().reset_index().head(10)
d_listing = d_listing.rename(columns = {'host_name_x' : 'Host Name','id' : 'Count'})
d_listing
```

```python
plt.figure(figsize = (8,4))
sb.barplot("Count", y="Host Name", data=d_listing ,palette="Blues_d")
plt.title("Top 10 Hosts")
```

```python
# Number of incorrect locations

e_listing = listings_merge['is_location_exact'].value_counts().to_frame().reset_index()
e_listing['index'] = e_listing['index'].replace('f','No').replace('t', 'Yes')
e_listing
```

```python
sb.barplot("is_location_exact", y = "index", data = e_listing ,palette = "muted")
plt.ylabel('Exact Location (Yes/No)')
plt.xlabel('Count')
plt.title("Exact Location (Yes/No)")
```

```python
# Which property type have how many listings

f_listing = listings_merge.groupby(['property_type'])['id'].count().sort_values().to_frame().reset_index()
f_listing
```

```python
sb.set(rc = {'figure.figsize':(11.7,8.27)})
sb.barplot("id", y = "property_type", data = f_listing, palette = "muted")
plt.ylabel('Count')
plt.xlabel('Property Type')
plt.title("Property wise Listing")
```

```python
# Number of listings for each Room Type 

g_listing = listings_merge.groupby(['room_type_x'])['id'].count().to_frame().reset_index()
g_listing
```

```python
sb.set(rc = {'figure.figsize':(7,5)})
sb.barplot("id", y = "room_type_x", data = g_listing, palette = "BuGn_r")
plt.ylabel('Count')
plt.xlabel('Room Type')
plt.title("Room wise Listing")
```

```python
# Number of accomodations in each property

sb.set(rc = {'figure.figsize':(12,9)})
sb.barplot("accommodates", y = "property_type", data = listings_merge, palette = "muted")
plt.title("Number of accomodations in each property")
```

```python
# Number of bathrooms in each property.

sb.set(rc = {'figure.figsize':(12,9)})
sb.barplot("bathrooms", y = "property_type", data = listings_merge, palette = "dark")
plt.title("bathrooms in each property")
```

```python
# Number of bedrooms in each property.

sb.set(rc = {'figure.figsize':(12,9)})
sb.barplot("bedrooms", y = "property_type", data = listings_merge, palette = "colorblind")
plt.title("bedrooms in each property")
```

```python
# Average price (in Dollars) of each property 

i_listing = listings_merge.groupby(['property_type'])['price_y'].mean().sort_values(ascending=False).reset_index()
i_listing
```

```python
sb.barplot("price_y", y = "property_type", data = i_listing, palette = "Blues_d")
plt.title("Average price (in Dollars) of each property ")
```

```python
# Average Rating of each property type

j_listing = listings_merge.groupby(['property_type'])['review_scores_rating'].mean().sort_values(ascending=False).to_frame()
j_listing = j_listing[j_listing.review_scores_rating.notnull()].reset_index()
j_listing
```

```python
# Number of properties lie in each cancellation policy.

k_listing = listings_merge.groupby(['cancellation_policy'])['property_type'].count().reset_index()
k_listing
```

```python
explode = (0,0,0,0,0,0)
plt.pie(k_listing['property_type'], labels = k_listing['cancellation_policy'], autopct = '%.2f%%', pctdistance = 0.8, explode = explode)
centre_circle = plt.Circle((0,0),0.50, fc = 'white')

# gcf = get current figure
# gca = get current axis

fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.title("Cancellation Policy")

plt.show()
```
